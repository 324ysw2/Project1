```python
# 선물, 현물 데이터 불러오기
import pandas as pd
FData = pd.read_csv("../dataset/kospi200_F202403.csv", index_col=0)
PData = pd.read_csv("../dataset/kospi200_15min.csv", index_col=0)
```

```python
# 선물 데이터 확인 및 전처리 계획
    # 음수 -> 양수로
    # drop
    # 컬럼명 변경
FData.head(5)
```

```python
FData = FData.abs()
FData
```

```python
# FData.drop('Unnamed: 0',axis = 1,inplace=True)
```

```python
FData = FData.loc[:,['체결시간','시가','고가','저가']]
```

```python
FData.columns = ['체결시간','선물_시가','선물_고가','선물_저가']
```

```python
FData.columns
```

```python
# 현물 데이터 확인 및 전처리 계획
    # 음수 -> 양수
    # 100으로 나눠주기
    # drop
PData.head(5)
```

```python
PData = PData.abs()
```

```python
cols_to_divide = ['현재가','시가','고가','저가']
PData[cols_to_divide] = PData[cols_to_divide] / 100
```

```python
# PData.drop('Unnamed: 0',axis = 1,inplace=True)
```

```python
PData = PData.loc[:,['체결시간','시가','고가','저가']]
PData
```

```python
PData.columns = ['체결시간','현물_시가','현물_고가','현물_저가']
```

```python
PData.columns
```

```python
# 데이터 합치기
Data = pd.merge(FData,PData,how='inner',on='체결시간')
```

```python
Data.head()
```

```python
# basis 계산
Data['basis_시가'] = Data['선물_시가']-Data['현물_시가']
```

```python
Data.head()
```

```python
Data = Data[['체결시간','basis_시가']]
```

```python
Data.to_csv('basis_2403.csv', index=False)
```

### 베이시스=선물 가격−현물 가격
* 양수인 경우 (콘탱고, Contango):
    - 의미: 선물 가격이 현물 가격보다 높은 상태를 가리킵니다.
    - 원인: 여러 요인이 있을 수 있으나, 주로 보관 비용, 금리, 보험료 등의 요소가 포함될 수 있습니다. 예를 들어, 선물 계약의 만기까지 원자재를 저장하는 데 발생하는 비용이나 금융 비용이 선물 가격에 반영되어 현물 가격보다 높아질 수 있습니다.
    - 특징: 콘탱고 상태에서는 선물 계약의 만기가 가까워질수록 베이시스는 일반적으로 감소하며, 만기일에는 선물 가격과 현물 가격이 수렴하게 됩니다.
* 음수인 경우 (백워데이션, Backwardation):
    - 의미: 선물 가격이 현물 가격보다 낮은 상태를 가리킵니다.
    - 원인: 수급 불균형, 특정 상품에 대한 급격한 수요 증가, 저장 공간의 부족 등 다양한 요인이 원인이 될 수 있습니다. 백워데이션은 특정 상품에 대한 현재의 실제 수요가 미래의 예상 수요보다 높을 때 발생하기도 합니다.
    - 특징: 백워데이션 상태에서는 선물 계약의 만기가 가까워질수록 선물 가격과 현물 가격은 수렴하게 됩니다.
