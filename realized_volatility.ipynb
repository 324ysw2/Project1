```python
ks200 = kospi_data[['Close', 'Daily_Return']]
daily_return = ks200['Daily_Return']
daily_return
```

```python
mean_list = []
k = 20
for i in range(1, len(ks200) + 1):
	sample_list = []
	sample_list = ks200['Daily_Return'][i:i+k]
	print(sample_list)
	sample_mean = round(np.mean(sample_list), 5)
	mean_list.append(sample_mean)
print(mean_list)
```

```python
ks200['sample_mean'] = mean_list
```

```python
ks200['sample_mean'] = ks200['sample_mean'].shift(1)
```

```python
ks200 = ks200.dropna()
ks200
```

```python
realized_volatility = []
vol__2 = []
k = 20

for i in range(1, len(ks200['Close'])):
    result = 0
    print(i)
    print("==================================================================")
    for j in range(1, k):
        try:
            log_returns = np.log(np.array(ks200['Close'][i+j-2 : i+j+k-2])) - np.log(np.array(ks200['Close'][i+j-1 : i+j+k-1]))
            sigma = log_returns - ks200['sample_mean'][i-1]
            result += sigma**2
            result = result / (k - 1)
        except:
            continue
    print(result)
```
