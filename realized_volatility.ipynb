```python
realized_volatility = []

for i in range(1, len(ks200['Close']) + 1):
    log_returns = np.log(np.array(ks200['Close'][1:])) - np.log(np.array(ks200['Close'][:-1]))
    rv = np.sqrt(252 * np.sum(log_returns**2) / len(ks200['Close']))
    realized_volatility.append(rv)

print(realized_volatility)
```
